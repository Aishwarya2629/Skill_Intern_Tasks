**IMPORTING MODULES & DATASET**

```python
import pandas as pd
import numpy as np  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df
```

**VISUALIZATION**

```python
# Visualizing the data
df.plot(x='Hours', y='Scores', style='*')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# Preparing the data
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

This line extracts the feature variables (X) from the data DataFrame. iloc[:, :-1] selects all rows and all columns except for the last column. .values converts the selected data into a NumPy array. In simpler terms, X will contain all the columns of the data DataFrame except the last one, assuming the last column is the target variable.

y = data.iloc[:, 1].values: 

This line extracts the target variable (y) from the data DataFrame. iloc[:, 1] selects all rows from the second column. .values converts the selected data into a NumPy array. In this case, y will contain the values of the second column of the data DataFrame, assuming the second column is the target variable.

So, in summary:

X contains the feature variables used for prediction.
y contains the target variable (in this case, likely the scores) that the model will try to predict based on the features in X.

**SPLITTING**

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**TRAINING**

```python
# Training the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Plotting the regression line
line = model.coef_ * X + model.intercept_
```

```python
# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line)
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

**PREDICTION**

```python
# Making predictions
hours = 9.25
predicted_score = model.predict([[hours]])
print("Predicted Score if a student studies for 9.25 hrs/day:", predicted_score[0])
```

```python
# Evaluating the model
y_pred = model.predict(X_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```

**CONCLUSION**

In conclusion, we find that if a person studies for 9.25 hours a day, he/she would be able to score around 93.69
