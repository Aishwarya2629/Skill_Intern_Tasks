**IMPORTING MODULES & DATASET**

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
```

```python
# Reading data from remote link
df = pd.read_csv("iris.csv")
df
```

```python
# Preparing the data
X = df.iloc[:, [0, 1, 2, 3]].values
```

```python
# Finding the optimum number of clusters using the elbow method for K-means
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

**ELBOW METHOD**

```python
# Plotting the elbow method graph for K-means
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for K-means')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
```

**PREDICTION**

We see that at the elbow occurs at 3. Thus, the optimal number of clusters is for the given iris dataset.

**KNN**

```python
# Applying k-means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualizing the clusters for K-means
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 2)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
```

**CONCLUSION**

we find that the most optimal number of clusters that can be formed is "3" by using the "elbow method"
